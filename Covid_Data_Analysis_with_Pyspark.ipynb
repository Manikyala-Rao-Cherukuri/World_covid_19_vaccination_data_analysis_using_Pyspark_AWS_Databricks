```python
# pyspark function
from pyspark.sql.functions import *
# URL processing
import urllib
```

```python
# Define file type
file_type = "csv"

# whether the file has a header
first_row_is_header = "true"

# Delimiter used in the file
delimiter = ","

# read csv file to spark dataframe
aws_keys_df=spark.read.format(file_type).option("header",first_row_is_header).option("sep",delimiter).load("/FileStore/tables/Databricks_user_accessKeys_updated.csv")
```

```python
# get the AWS access and secret key from AWS dataframe
ACCESS_KEY = aws_keys_df.where(col('User name') == 'Databricks_user').select('Access key ID').collect()[0]['Access key ID']
SECRET_KEY = aws_keys_df.where(col('User name') == 'Databricks_user').select('Secret access key').collect()[0]['Secret access key']

# Encode the secret key
ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe="")
```

```python
# AWS S3 bucket name
AWS_S3_BUCKET = "world-covid-dataset"

# Mount name for the bucket
MOUNT_NAME = "/mnt/world-covid-dataset"

# Source URL
SOURCE_URL = "s3n://{0}:{1}@{2}".format(ACCESS_KEY, ENCODED_SECRET_KEY, AWS_S3_BUCKET)

# Mount the driver
dbutils.fs.mount(SOURCE_URL, MOUNT_NAME)
```

```python
# trying to fetch the details from S3
%fs ls /mnt/world-covid-dataset/
```

```python
# creating the pyspark dataframe, after reading the data from S3

# File Location and type
file_location = "/mnt/world-covid-dataset/country_vaccinations.csv"
file_type = "csv"

#CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df_pyspark = spark.read.format(file_type) \
	.option("inferSchema",infer_schema,) \
	.option("header",first_row_is_header) \
	.option("sep", delimiter) \
	.load(file_location)
# displaying the data
display(df_pyspark)
```

```python
df_pyspark.printSchema()
```

```python
df_pyspark.createOrReplaceTempView('covid_data')
```

```python
# command to used to save the spark dataframe into permanent table in database
df_pyspark.write.mode("overwrite").saveAsTable("coviddata")
```

```python
df_pyspark.select("*").show(truncate=False)
```

```python
# creating custom pyspark dataframe
df_pyspark_m = df_pyspark.select(['country','date','total_vaccinations','people_fully_vaccinated','daily_vaccinations','daily_vaccinations_per_million'])
display(df_pyspark_m)
```

```python
# renaming the cloumn name
df_pyspark_m = df_pyspark_m.withColumnRenamed('total_vaccinations','total_people_vaccinated_per_day' )
display(df_pyspark_m)
```

```python
# calculating the sum of total people vaccinated with respective to the country
result = df_pyspark_m.groupBy('country').sum('total_people_vaccinated_per_day')
display(result)
```

```python
# calculating the Top 10 countries having high daily_vaccination sum
result = df_pyspark_m.groupBy('country').sum('daily_vaccinations').orderBy(desc('sum(daily_vaccinations)')).limit(10)
display(result)# calculating the Top 10 countries having high daily_vaccination sum
result = df_pyspark_m.groupBy('country').sum('daily_vaccinations').orderBy(desc('sum(daily_vaccinations)')).limit(10)
display(result)
```

```python
# Top cuntries, who crossed daily more than 1 million vaccinations
grouped_result = df_pyspark_m.groupBy('country').agg(sum('daily_vaccinations_per_million').alias("sum_of_vaccinations")).orderBy(desc('sum_of_vaccinations'))
result =  grouped_result.filter(col('sum_of_vaccinations') >= 1000000).orderBy(desc('sum_of_vaccinations')).limit(10)
display(result)
```
